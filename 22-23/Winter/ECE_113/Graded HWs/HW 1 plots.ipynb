```python
# imports
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def x(t, f0):
    return np.cos(2*np.pi*f0*t)
```

```python
# i. $f_0=3$ Hz, $f_s=10$ Hz
t = np.linspace(0, 2, 1000)
f0 = 3
fs = 10
plt.plot(t, x(t, f0))
plt.stem(np.arange(0, 2 + 1/fs, 1/fs), x(np.arange(0, 2 + 1/fs, 1/fs), f0), 'r')
plt.title('$f_0=3$ Hz, $f_s=10$ Hz')
plt.show()
```

i) $F_0$= 3Hz

$2\cdot F_0 = 6Hz \le f_s = 10Hz$


$\therefore$ We should be able to recover the signal since we are above the Nyquist frequency.
