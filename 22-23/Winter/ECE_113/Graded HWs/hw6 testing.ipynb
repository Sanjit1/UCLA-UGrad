```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lets say we have x[n] = (N/2) 1s followed by (N/2) 0s
N = 10
x = [1] * (N//2) + [0] * (N//2)
# lets convolution this with itself
y = np.convolve(x, x)
# lets plot the result
plt.stem(y)
plt.show()
```

```python
# Lets plot x+1 for n = 0 to (N/2)-1 and N-1-x for n = (N/2) to N-1
x1 = [x+1 for x in range(N//2 -1)] + [(N/2)-x for x in range(N//2)]
plt.stem(x1)
plt.show()
```

```python
# Take and plot the DFT of x1
X = np.fft.fft(x)
# Plot real and imaginary parts of X1 with labels and legends and titles
plt.stem(np.real(X))
plt.stem(np.imag(X), 'r')
plt.title('Real and Imaginary parts of X')
plt.xlabel('n')
plt.ylabel('X[n]')
plt.legend(['Real', 'Imaginary'])
plt.show()
print('Real part of X = ', np.real(X))
print('Imaginary part of X = ', np.imag(X))
```

```python
# Lets plot phase and magnitude of X1
plt.stem(np.angle(X))
plt.stem(np.abs(X), 'r')
plt.title('Phase and Magnitude of X')
plt.xlabel('n')
plt.ylabel('X[n]')
plt.legend(['Phase', 'Magnitude'])
plt.show()
print('Phase of X = ', np.angle(X))
print('Magnitude of X = ', np.abs(X))
```
