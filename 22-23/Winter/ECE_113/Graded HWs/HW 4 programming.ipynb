```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we have two signals
x = np.array([0, 2, 3, -3, 4, 1])
h = np.array([-3 ,5 ,-6 ,4 , 0])
# Convolution 
y = np.convolve(x,h)
# plot
plt.stem(y)
print(y)
plt.show()
```

```python
# np.convolve takes too long
# We just want to evaluate the convolution at one point
# We can do this by using the dot product
# Lets evaluate the convolution at index 1
L = len(x)
M = len(h)
y = [0]*(L+M-1)
evaluated_index = 2
for i in range(L):
    if(evaluated_index-i>=0):
        print(i,evaluated_index-i)
        y[evaluated_index] += x[i]*h[evaluated_index-i]
print(y)
plt.stem(y)
plt.show()



```
