```python
X = np.random.normal(0, 1, 1000) # X = ζ(0,1)
Y = 5*X + 0 # Y = aX + b, a=5, b=0
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y=aX+b')
plt.title('Scatter plot of X and Y, a=5, b=0')
plt.show()
```

```python
X = np.random.normal(0, 1, 1000) # X = ζ(0,1)
Y = -4*X + 0 # Y = aX + b, a=-4, b=0
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y=aX+b')
plt.title('Scatter plot of X and Y, a=-4, b=0')
plt.show()
```

```python
X = np.random.normal(0, 1, 1000) # X = ζ(0,1)
Z = X**2 # Z = X^2
plt.scatter(X, Z)
plt.xlabel('X')
plt.ylabel('Z=X^2')
plt.title('Scatter plot of X and Z')
plt.show()
```

```python
X = np.random.normal(0, 1, 1000) # X = ζ(0,1)
W = X**3 # W = X^3
plt.scatter(X, W)
plt.xlabel('X')
plt.ylabel('W=X^3')
plt.title('Scatter plot of X and W')
plt.show()
```
