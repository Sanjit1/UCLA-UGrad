```python
import numpy as np
import matplotlib.pyplot as plt
def charact(w, pdf, range):
    x = np.linspace(range[0], range[1], 10000)
    return np.sum(pdf(x)*np.exp(1j*w*x))/10000


def plot_char(pdf, range, name):
    char = lambda w: charact(w, pdf, range)
    # plot the characteristic function for w between -10 and 10 
    ws = np.linspace(-10,10,10000)
    plt.subplot(2,1,1)
    plt.plot(ws, np.real([char(w) for w in ws]))
    plt.title('Real part of the characteristic function of ' + name)
    plt.xlabel('w')
    plt.ylabel('Re{G(w)}')
    plt.subplot(2,1,2)
    plt.plot(ws, np.imag([char(w) for w in ws]))
    plt.title('Imaginary part of the characteristic function of ' + name)
    plt.xlabel('w')
    plt.ylabel('Im{G(w)}')
    plt.show()
```

```python
def gaussian(x):
    return np.exp(-x**2/2)/np.sqrt(2*np.pi)

plot_char(gaussian,[-10000, 10000],'Gaussian')

# Characteristic function of a random variable X with a uniform distribution
def uniform(x):
    return 1*(np.abs(x)<=1)

plot_char(uniform, [-50, 50], 'Uniform')

# Characteristic function of a random variable X with a exponential distribution
def exponential(x):
    return 1*(x>=0)*np.exp(-x)

plot_char(exponential, [-100, 1000], 'Exponential')
```

```python
# Characteristic function of a random variable X with a chi-squared distribution
def chi_squared(x):
    return 1*(x>=0)*np.exp(-x/2)

plot_char(chi_squared, [-100, 1000], 'Chi-squared')
```

```python
# Characteristic function of a random variable X with a semi-circle distribution
def semi_circle(x):
    return 1*(np.abs(x)<=1)*np.sqrt(np.abs(1-x**2))

plot_char(semi_circle, [-500, 500], 'Semi-circle')
# K now im just fucking around lmfao
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcdefaults()
```

```python
# Two RVs with uniform distributions X and Y
# Lets Sample rvX and rvY 1000 times each and plot them one by one
sampledX = np.random.uniform(0,1,100000)
sampledY = np.random.uniform(0,1,100000)
# Plot histogram of sampledX
plt.subplot(2,1,1)
plt.hist(sampledX, bins=100)
plt.title('Histogram of sampled X')
plt.xlabel('X')
plt.ylabel('Frequency')
# Plot histogram of sampledY
plt.subplot(2,1,2)
plt.hist(sampledY, bins=100)
plt.title('Histogram of sampled Y')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.show()
```

```python
# Lets plot the sum of X and Y.
sampledA = sampledX + sampledY
plt.hist(sampledA, bins=100)
plt.title('Histogram of sampled Z')
plt.xlabel('Z')
plt.ylabel('Frequency')
plt.show()
```

```python
# Lets plot X-Y
sampledB = sampledX - sampledY
plt.hist(sampledB, bins=100)
plt.title('Histogram of sampled W')
plt.xlabel('W')
plt.ylabel('Frequency')
plt.show()
```

```python
# Lets plot the product of X and Y
sampledC = sampledX * sampledY
# Lets plot the histogram of sampledC as a distribution aka have % of values in each bin
plt.hist(sampledC, bins=100, density=True)
plt.title('Histogram of sampled V')
plt.xlabel('V')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Lets play with CDFs
x_cdf = np.sort(sampledX)
x_cdf_ls = np.linspace(0,1,100000)
y_cdf = np.sort(sampledY)
y_cdf_ls = np.linspace(0,1,100000)
# Now lets take the convolution of the CDFs
z_cdf = np.convolve(x_cdf, y_cdf)
z_cdf_ls = np.linspace(0,1,200000-1)
# Plot the CDFs
plt.subplot(2,1,1)
plt.plot(x_cdf, x_cdf_ls)
plt.title('CDF of X')
plt.xlabel('X')
plt.ylabel('CDF')
plt.subplot(2,1,2)
plt.plot(y_cdf, y_cdf_ls)
plt.title('CDF of Y')
plt.xlabel('Y')
plt.ylabel('CDF')
plt.show()
# Plot the convolution of the CDFs
plt.plot(z_cdf, z_cdf_ls)
plt.title('CDF of Z')
plt.xlabel('Z')
plt.ylabel('CDF')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate the tossing of 100 coins 1000 times and plot the histogram of the number of heads
# Highlight the area between 40 and 60 and calculate the estimated probability of getting between 40 and 60 heads
# Lets simulate the tossing of 100 coins 1000 times
sampledY = np.sum(np.random.randint(0,2,(100,100000)), axis=0)
# Lets plot the histogram of sampledY
plt.hist(sampledY, bins=100)
plt.title('Histogram of sampled Y')
plt.xlabel('Y')
plt.ylabel('Frequency')

# Lets highlight the area between 40 and 60
plt.hist(sampledY, bins=100)
plt.title('Histogram of sampled Y')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.axvspan(40, 60, alpha=0.5, color='red')
plt.show()
# Lets calculate the estimated probability of getting between 40 and 60 heads
print('Estimated probability of getting between 40 and 60 heads: ', np.sum((sampledY>=40) & (sampledY<=60))/100000)
# Calculate the mean and variance of the distribution
print('Mean of the distribution: ', np.mean(sampledY))
print('Variance of the distribution: ', np.var(sampledY))

```
