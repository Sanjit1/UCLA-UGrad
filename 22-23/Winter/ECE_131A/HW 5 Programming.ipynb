```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def M_cdf(x, p=0.6):
    return 1-(1-p)**x

def X_cdf(x, l=1):
    return 1-np.exp(-l*x)

k = np.arange(1, 11)
t = np.linspace(0, 10, 1000)

plt.figure(figsize=(10, 5))
plt.plot(k, 1-M_cdf(k), 'o', label='P(M>k)')
plt.plot(t, 1-X_cdf(t), label='P(X>t)')
plt.xlabel('k or t')
plt.ylabel('P(M>k) or P(X>t)')
plt.legend()
plt.show()
```
